```python
classifier = pipeline("sentiment-analysis")

classifier([
    
"""Executive Summary Areas hosting refugees and IDPs in the East and Horn of Africa and the Great Lakes region are beset by 
socioeconomic fragility and in some instances political volatility. Despite these challenges, States in the 
region continue to generously provide asylum, notably Uganda, which is likely to remain the largest 
refugee-hosting country in Africa in 2022. Sudan also hosts over 1 million refugees. Tragically, violence 
and insecurity may continue in the region, with risks of further displacement arising from conflict in 
northern Ethiopia, tensions along the Sudan-Ethiopia-Eritrea border, a fragile peace process in South 
Sudan, and political transition and inter-communal conflict in Sudan. Political complexities and ongoing 
conflict in Somalia, human rights violations in eastern Democratic Republic of the Congo, and post-crisis 
challenges in Burundi also present risks of new displacement in the region. Climate shocks, food 
insecurity and COVID-19 are likely to further deepen the plight of those forced to flee.
COVID-19 remains a major focus, having already harmed the livelihoods, health and well-being of 
refugees and IDPs in the region. High-frequency phone surveys –  conducted with the World Bank and 
national statistics offices in Djibouti, Ethiopia, Kenya, Rwanda and Uganda –  indicate the pandemic has 
reduced incomes, including from remittances, and access to work, while food prices have increased. 
Displaced populations were driven towards negative coping mechanisms such as reducing food 
consumption, depleting savings and selling assets.  It will be critical to strengthen support for 
livelihoods to mitigate the socioeconomic impact of COVID-19 in 2022.
 UNHCR’ s protection strategy in 2022 will prioritize advocacy and dialogue with authorities, aiming to 
maintain access to asylum, strengthen asylum systems and uphold the principle of non-refoulement. 
Systems for gender-based violence prevention and response, child protection and community-based 
protection structures will be strengthened. Efforts will be made to address the root causes of onward 
movement, with exercises to map protection services available for vulnerable people on the move 
towards North Africa, Europe and the Gulf.Donor support will be vital to protect forcibly displaced people; to assist them, empower them and help 
find solutions to their situations and build better futures, as well as to the communities that host them. 
Significant resources will be required for even the most basic needs of people affected by the 
displacement situations in the region. Protection, health, education, livelihoods, water, sanitation and 
hygiene will continue to be provided and strengthened. Predictable donor support will be critical to 
fulfilling the promise of international responsibility-sharing at the heart of the Global Compact on 
Refugees.The region holds promise for solutions in countries in an early recovery or post-conflict phase. The IGAD 
Solutions Initiative for South Sudan and Sudan focuses on national strategies and action plans for 
solving the displacement crises affecting both countries. Consultative processes seek to identify points 
of focus like integrated area-based programmes for refugees, IDPs, returnees and hosting communities, 
and support to Global Refugee Forum (GRF) policy pledges made by governments. South Sudanese 
refugees are expected to continue returning spontaneously to areas experiencing relative peace. Places 
of possible return will require significant funding by both humanitarian and development partners. 
New and continuing strategic partnerships will focus on realizing the potential of the Global Compact 
and on GRF pledges, as well as advancing self-reliance among people of concern by ensuring their 
inclusion in host communities and national systems.
In addition to traditional partners, UNHCR will seek to cooperate with development actors, financial 
institutions and civil society to achieve these goals. As part of the SDG process, UNHCR and IOM co-lead 
the Opportunities/Issues-Based Coalition on Forced Displacement, an opportunity to strengthen the 
humanitarian-development nexus and increase attention on mixed movements.
Section 1: Strategic Direction1.1 - Context (HQ/Bureau)
1.General Operational ContextThe operational environment in the 11 countries of the East, Horn and Great Lakes region of Africa is characterized by 
heightened socio-economic fragility and political volatility affecting areas hosting refugees and internally displaced persons. The region is host to 4.6 million refugees originating in large part from South Sudan, the Democratic Republic of Congo, 
Somalia, Sudan, Ethiopia and Eritrea.   
With 8 million IDPs, the sub region has witnessed a sharp increase of over 1 million internally displaced persons as 
compared to 2020. The trend of increase is attributable to a combination of factors which include protracted and new 
forced displacement crises fueled by ongoing sub-national armed inter-communal conflicts, climatic shocks, and socio-
economic fragility.
Successive waves of refugee influxes and internal displacement have created unprecedented pressure on the limited 
economic resources and basic services available for countries in the region.  For 2022, Sudan’ s and South Sudan’ s 
displacement situations are expected to be largest displacement producing situations whereas Uganda will likely remain 
the largest refugee hosting country in Africa. The region also holds promise for solutions to displacement in countries 
which are either in an early recovery or post conflict phase. Predictable donor support to protection and solutions for 
refugees, IDPs, returnees and host communities outlined in this 2020 interim strategy will be critical to harness these 
opportunities and fulfill the promise of international responsibility sharing at the heart of the Global Compact on Refugees 
(GCR).
1.	Root Causes of Forced Displacement
i) Multiple Factors of Fragility: 
Climate Shocks and Food Insecurity. Climatic shocks (prolonged droughts, extreme flooding, and desert locust 
infestations) are factors triggering new displacement as well as impacting those already displaced. Throughout the 
region, climatic shocks are commonplace in the arid and semi-arid refugee hosting areas which cover 60-70 percent of 
the region. The region is facing critical levels of nutrition and food insecurity hosting three out of the ten worst food crises 
in the world for the past three years 2018-2020 (South Sudan, Ethiopia, and Sudan). Dependency of the refugee 
population on humanitarian food and non-food assistance is increasing. With limited livelihood and agriculture 
opportunities, in addition to successive and deep food ration cuts of 23-60% affecting over 3.4 million or 74% of the total 
refugee population, 7 of the 11 country operations are experiencing climate and economic shocks, compounded by COVID-19 
related measures. These cumulative factors are anticipated to have   severe negative and multifaceted impacts, exacerbating 
underlying poor nutritional and economic status of refugee populations across the region in 2022. Addressing food and 
nutrition needs will remain of the utmost priority primarily due to the combined impact of on-going armed conflicts and 
climate shocks. The rapid loss of sources of livelihoods has also alarmingly increased suicide rates among refugee youth, a 
very large segment of the population of concern in the region. Through protection monitoring data, a direct correlation could 
be established between the reduction of humanitarian assistance levels and the rise in sexual and gender-based violence in 
the region.
Socio-Economic Impact of COVID-19. UNDP Africa’ s new report: “ Analyzing long-term socio-economic 
impacts of COVID-19 across diverse African contexts”  examines the effect COVID-19 will have on macro-economic 
development in 10 African countries by 2030 and 2050. The study highlights the multifaceted economic and social 
consequences of the COVID-19 pandemic across Africa, including on human development indicators, economic 
interdependence, and growth and resilience patterns. The level of household impoverishment is likely to be heightened 
among refugee households and to increase further in 2022 as second and third waves of COVID-19 strike the people of 
the region whose resilience is eroded. The COVID-19 containment measures which include closure of markets have 
short, medium- and long-term adverse consequences on the resilience and coping mechanisms of refugees. The Bureau 
will focus on mitigating the adverse socio-economic impacts of the COVID-19 pandemic which are expected to increase 
poverty, shrink economic growth rates, and redouble lifesaving needs of 4.6 million refugees and asylum-seekers and over 8 
million IDPs. The roll out of vaccination and socio-economic recovery programmes through the COVAX facility and other 
CDC Africa initiatives to boost local production of vaccines are not expected to reach all refugees, IDP and returnee 
areas in the region. Direct and indirect effects of the ongoing COVID-19 pandemic will continue to overstretch health 
systems, and global shortages of essential medical supplies will negatively affect regular health services. Additionally, 
increasing infections among healthcare workers will continue to pose a significant challenge. While there has so far been no 
large-scale outbreak of COVID-19 among PoCs, this will remain a risk in 2022. All countries in the region currently have 
widespread community transmission and have reported cases among PoCs.   The water and sanitation response will also 
need to be maintained of scaled up to avert the risk of rapidly rising COVID-19 rates including of new strains in refugee, IDP 
and returnee settings.
ii) Peace and Security Related Triggers of Forced Displacement: Regional threats to peace and stability 
that could trigger forced displacement have seen an increase in 2021 which could continue in 2022.
Ethiopia situation: The situation in Tigray remains dire, characterized by human rights violations and growing 
unattended humanitarian needs of an estimated 500,000 IDPs who are predominantly ethnic Tigrayan. As restrictions on 
access are gradually lifted in 2021, the full extent of needs in Tigray will necessitate a whole of UN approach to scale up 
the IDP response. This includes increasing support to the CCCM, shelter and protection cluster. Similarly, in the 
Benishangul Gumuz region over 100,000 IDPs need assistance and access is restricted. Access to Eritrean refugees in 
Ethiopia remains volatile and Eritrean refugees will continue to face heightened legal and protection needs in Ethiopia. 
New influxes of South Sudanese in Gambella and a potential influx from Somalia make Ethiopia the most complex and 
the least predictable emergency in the region where the needs are growing the fastest.
 
The national elections in Ethiopia to be held on 5 June 2021 will be a watershed moment for the Horn of Africa region. 
Ethiopia’ s elections, whether they are held on time or postponed, could spark clashes between perceived winners and 
losers at the polls. While the Ethiopian Prosperity Party is expected to win, the main opposition groups such as the 
Oromo Federalist Party are boycotting the elections. In the aftermath of the elections, an increase of both internal 
displacement and refugee outflows from Ethiopia may occur if efforts at intercommunal dialogue do not restore social 
cohesion in several regional states.
Sudan -Ethiopia- Eritrea Border Tensions: Ethiopia’ s filling of the Grand Ethiopian Renaissance Dam (GERD) 
scheduled in July 2021 could be another major flashpoint. Ethiopia, Sudan, and Eritrea have amassed troops in the al-
Fashaga triangle placing these three countries on the brink of an open regional conflict. The role and geopolitical 
interests of States such as Egypt, the United Arab Emirates and the Kingdom of Saudi-Arabia and their proxies pose the 
added risk of internationalizing the conflict with devasting consequences in terms of forced displacement and regional 
peace and stability. In 2021, refugees have been the collateral or direct victims of armed attacks by State and non-state 
actors.
South Sudan:  The South Sudan displacement situation accounts for 2.2 million refugees and remains the largest 
scale forced displacement crisis in Africa. It is third largest refugee population globally after Syria’ s and Afghanistan’ 
s,  yet it is among the most poorly funded displacement situations.  While important steps were taken to advance the peace 
process these have yet to yield tangible peace dividends for the displaced.  A modicum of the Rule of law has also not been 
restored in most parts of the country. Violations of human rights and humanitarian law including the use of rape as a weapon 
of war are being committed with near complete impunity by all warring parties. Against this backdrop, Upper Nile and 
Equatoria states have become flash points for the resurgence of sub-national conflicts between clans and sub-clans forcing 
these populations into displacement throughout 2021. This trend will likely continue in 2022 unless underlying causes linked 
to power sharing among ethnic are resolved at the grass roots community level. UNMISS Protection of Civilian (PoC) sites 
were transformed into IDPs camps in 2021 but no commensurate resources have been allocated by humanitarian and 
development actors to support reintegration of IDPs. Dwindling humanitarian assistance and support to return areas makes 
return unstainable and pendular movements of refugees. Additionally, the sharp rise in food insecurity in parts of South 
Sudan have caused persons to flee to Sudan’ s White Nile State anew due to a mix of reasons including conflict, and climate 
induced food insecurity. The drastic food ration cuts experienced by refugees in South Sudan in 2021 can flare tensions 
between 300,000 refugees, IDPs and returnee population in 2022.  These factors call for multilateral donors to urgently 
reprioritize funding to South Sudan’ s displaced both within and outside the country .    
Burundi situation:  The human rights situation in Burundi has markedly not improved according to all available 
indicators. The practice of enforced disappearances, extrajudicial arrests, and killings by the Imbonerakure (youth militia) 
are frequently reported. Despite these trends, the Government has shown a growing disposition to work with the 
international community to address the COVID-19 pandemic and create more conducive conditions for return. Despite 
existing hurdles to reintegration over 140,000 Burundian refugees have voluntarily repatriated from throughout the region. 
This surge in returnees has added strain to overstretched public services and access to arable land in one of the most densely 
and poorest countries in the world.  The political overtures by the Government of Burundi which has welcomed its citizens 
will be important to capitalize upon in 2022 through increased resource mobilization efforts to support reintegration. 
Concurrently, Tanzania’ s protection environment has continued to deteriorate for both Burundian and other refugees. 
Maintaining protection space in Tanzania will therefore remain paramount in light of the recent political changes unfolding in 
the country, UNHCR might have new opportunities to engage.
Somalia situation: The latent dispute between Jubaland and the Federal Government of Somalia remains largely 
unaddressed and the date of Somalia’ s elections and its outcomes are uncertain. Casualties and forced displacement 
have already occurred in the Gedo border region between Kenya and Somalia. Somalia and Kenya are also embroiled in 
a maritime dispute which shows no signs of being resolved by the ICJ.  The bilateral tensions between Kenya and Somalia 
may have prompted Kenya to restrict asylum for refugees in Dadaab and Kakuma.  The Government of Kenya has reiterated
its request for the camps to be closed in March 2021.The resurgence of Al-Shabab and other armed group also 
accentuates the risk of forced recruitment of the refugee and IDP youth into armed groups. The civilian and humanitarian 
character of asylum has also been threatened by the rise of non-state armed groups. In light of the continued downsizing 
of UN and AU peacekeeping missions in Somalia, Sudan, and South Sudan, UNHCR will be expected to invest into and 
spend considerable resources on the relocation of refugees and IDPs to safe sites in 2022. The threats to protection that 
will continue to prevail in 2022 include: cross-border attacks on displacement sites; cattle rustling; clashes among 
different displaced communities of the same nationality and increasing rivalries between refugee, returnee and host 
communities over scarcer natural resources and access to housing land and property rights.
Overall, the aftermath of the conflict in Tigray region of Ethiopia and its ramifications on the country’ s and the region’ s 
stability; the situation in the Sudans; the ongoing dispute between Somalia and Kenya and the rise of human rights 
violations in Ituri and the Kivus in the eastern DRC are not expected to get resolved in 2021. This context makes 
sizeable return of refugees from these areas of origin a very unlikely prospect for 2022. The fall out of the forthcoming 
elections in Ethiopia to be held in June 2021, political instability, and conflict in advance of the forthcoming elections in 
the Democratic Republic of Congo (DRC) in 2023 may result in refugee outfluxes into Burundi, Rwanda, and Tanzania. 
The security situation in Somalia, Ethiopia and Eritrea also remains fragile, and it is a source of political tension and 
trigger of internal and refugee forced displacements within the region in 2022.
 B. Regional Protection and Solution Context
i) Access to and Quality of Asylum: Prior to the outbreak of COVID-19 in March 2020,  observable trends included 
shrinkage in the asylum space, limitations imposed by some States to UNHCR’ s supervisory role and a general tendency on 
the part of governments not to portray asylum (or asylum capacity development) as a key priority. During the first quarter of 
2021, there were no indications of a marked improvement in the protection environment in the region. A deterioration was 
noted with incidents of refoulement and or denial to access to asylum procedures in Tanzania, Uganda, and Rwanda. There 
were reports of large-scale forced returns of Eritrean refugees from Ethiopia’ s Tigray region throughout the last quarter of 
2020 and the first quarter of 2021. Although these reports could not be verified by UNHCR Eritrea as it is not granted access 
to refugees or returnees, eyewitness testimonies by Eritrean refugees in Sudan and in the diaspora have corroborated these 
alleged acts of refoulement.  While limited improvements in humanitarian access did occur notably in Blue Nile State of 
Sudan, the overall planning assumption is that asylum space will continue to be challenging to secure in 2022.   
Access to territorial asylum and refugee status determination  procedures including registration of new arrivals, 
documentation and refugee status determination (RSD) continued to be hampered by COVID-19 mitigation measures, 
which, albeit gradually easing, will significantly continue to impede the arrival of new asylum-seekers in 2022. Cross-
border movements continue through unofficial border crossing points but continued formal border closures prevented 
border monitoring and complicated efforts to engage with border officials to ensure access to asylum and safeguards 
against refoulement in 2021. COVID-19 border surveillance measures have at times been used as a pretext to deny 
asylum to persons in need of international protection.
Strategic priorities will include addressing the root causes of onward movement, maintaining access to asylum/non-
refoulement, preservation of the civilian and humanitarian character of asylum. Emphasis will need to be placed on the 
need for asylum procedures in the region to be ‘ adaptable’  (e.g. to adjust to new influxes, to adjust procedures to permit 
continuation of asylum procedures despite COVID or other restrictions, and to prevent collapse of the procedures or 
severe long-term deterioration) while also ensuring integrity (noting, for example the increased potential for unforeseen 
procedural irregularities arising from ‘ working from home’ ) noting that the potential for extortion/other forms of corruption 
might be particularly of concern due to the COVID-induced financial desperation of host state populations, and/or in 
States with nascent asylum systems and/or where government staffing structures are not sufficiently robust to ensure 
effective supervisory accountability.  To date, limited progress has been seen in implementation of GRF pledges relating 
to strengthening the institutional asylum capacity. Kenya has moved ahead with discussions with Denmark on a possible 
‘ support request’  to be matched through the Asylum Capacity Support Group (ACSG). However, the pandemic and 
other factors/sensitivities have resulted in limited progress in terms of GRF pledge implementation by other States who 
made such pledges (notably Uganda, Rwanda, Ethiopia, Tanzania).
ii) Prospects for Solutions. Sudan and South Sudan’ s unfolding peace processes will need to be robustly 
supported to create conditions conducive for return for over 7 million refugees and IDPs of both countries. In 2022, the 
Sudan and South Sudan displacement situations will continue to be the largest in the region for the third consecutive 
year. 
In Sudan, while the Juba Peace Agreement was signed by most of the warring parties in October 2020, important 
armed factions remain outside the process in Darfur and the two areas of Blue Nile State and Kordofan. Limited access 
to the two areas where a vast majority of Sudanese refugees in Ethiopia and South Sudan originate will require UNHCR 
to scale up response to enable the implementation of solutions. Threats to physical and legal protection of refugees 
including of the civilian and humanitarian character of asylum are expected to remain rife in Eastern Sudan exacerbated 
by the latent border dispute with Ethiopia.  The complexity of developing a viable solutions framework for refugees in Sudan 
is that it must simultaneously address the needs of new and protracted situations. The newly arrived 60,000 Ethiopian 
refugees from Tigray and pre-existing Eritrean refugees face a precarious situation and will remain highly dependent on 
emergency assistance.  The new influx is competing for limited resources with displaced populations who have lived in 
Eastern Sudan for decades.  The urban refugee populations residing in the open areas of Khartoum as well as the new arrival 
of Congolese, Chadian and Central African refugees will also require sustained support as they have been negatively impacted 
by instability in Sudan. In 2022, the UNHCR operation in Sudan will be expected to require large scale resources devoted to 
relocation and site planning occurring simultaneously in Eastern and Blue Nile State, which has experienced a new refugee 
influx from Ethiopia’ s Benishangul Gumuz region since 2021. Against this backdrop, hyperinflation, coupled by a rising 
COVID-19 positivity rate and growing political rivalries within Sudan’ s transitional government combine to adversely 
impact the protection environment for refugees, IDPs and returnees in Sudan. 
In South Sudan, the transitional period charted by the Revitalized Agreement was extended until 2023.The lack of 
demobilization of armed forces and delays in the implementation of Chapter III of the Agreement on humanitarian 
assistance and reintegration are priorities that will need to be addressed in 2022 for South Sudanese to contemplate 
sustainable returns. Human rights concerns persist in Upper Nile and other States of South Sudan combined with food 
insecurity affecting 6 million South Sudanese. Therefore, the 2.2 million number of refugees is not expected to reduce in 
2022.   Concurrently, due to the relative return to peace in most parts of the country and significant reduction of 
humanitarian assistance available in asylum countries have prompted over 360,000 South Sudanese refugees to return 
spontaneously to South Sudan predominantly from Uganda and Sudan.  While UNHCR is not facilitating returns to South 
Sudan, a substantial increase to meet the growing reintegration needs and stabilize South Sudanese returnees will be needed 
in 2022. Lack of investment in South Sudan’ s early recovery need would result in refugee returnees swelling the ranks of the 
IDPs or engaging in dangerous cross-border pendular movements amidst the COVID-19 pandemic. The phenomena of 
pendular cross-border movements is attributable to returnees being unable to reintegrate.
Voluntary Repatriation and Reintegration of Burundians.   By 2022, nearly half of the remaining 300,000 
Burundian refugee population in the EHAGL are projected to opt to voluntarily  repatriate  mostly from Tanzania and 
Rwanda.  The reintegration needs of returnees in Burundi will double requiring increased investment in one of the 
poorest and most agrarian countries in Africa.  The most acute reintegration needs in Burundi will be livelihoods and 
shelter.
Repatriation of Ethiopian and Eritrean refugees is unlikely to increase given the prevailing humanitarian context in the 
Tigray region and its ripple effects.  Similarly, the massive human rights violations unfolding in the Eastern DRC’ s Ituri 
and Kivu provinces make prospects of voluntary repatriation to these countries dim. Local integration could be feasible 
for remaining small size refugee populations in protracted situations in the Great Lakes region. In 2022, the Bureau will 
pursue a study with the ICGLR which addresses risks of stateleneess and pave the way for a durable solutions strategy 
of the GL sub-region.  
2. Overall Coordination and Response:
In 2022, the Regional Bureau will continue to assert UNHCR’ s relevance through proactively engaging in humanitarian, 
development bilateral and multilateral mechanisms.
 
A. Coordination mechanisms
i) IDPs: UNHCR’ s scale-up of its IDP operations will be an operational priority for 2022. In line with the new IDP policy, 
Sudan, South Sudan, Ethiopia, and Somalia will be key operations where UNHCR assumes greater accountability for 
coordination and delivery of the response. UNHCR is increasing its footprint in geographical coverage, with a focus on 
emergency preparedness and pathways to transitional solutions. Burundi has a protracted IDP population whose needs will 
be addressed within the framework of the national reintegration strategy.
Through its leadership of the protection cluster, shelter and Camp Coordination and Management (CCCM) 
UNHCR resources will need to be increased to create a conducive environment for IDP returnees through the 
implementation of projects to prioritize fundamental protection and uphold the right of return of IDPs and their access to 
rights in areas of origin.
ii) Refugee Response and Reintegration. The Regional Refugee Response Plans (RRRPs) and Humanitarian 
Response Plans (HRPs) will remain key interagency coordination and response tools to address the lifesaving resilience 
and early recovery needs of refugees and IDPs in 2022. The needs of the South Sudanese RRRP and the HRPs exceed 
3 billion dollars for 2021. The needs for South Sudan’ s refugees, IDPs and returnees are expected to increase in 2022 
as a result of the socio-economic impact of COVID-19. Despite being the largest refugee situation in Africa the needs of 
South Sudanese have remained among the least funded globally.
The situation of Ethiopian refugees in Sudan will continue to be addressed  through a dedicated response in Sudan, 
whereas the RRRP for the Burundian situation may see a reduction as result of increased voluntary returns of Burundian 
refugees, but a commensurate increase in the needs of Burundian returnees (covered under the JRRRP) is expected in 
2022. This is justified by the fact that over one third of the 315,000 Burundian refugees will have availed themselves of 
facilitated voluntary repatriation primarily from Rwanda and Tanzania and other countries in the East, Great Lakes and 
Southern Africa. In 2022, the needs of Burundian refugee returnees will be addressed through the Joint Refugee Return 
and Reintegration Plan (JRRRP) endorsed by the Government of Burundi and co-led by UNHCR and UNDP. 
B. Strategic Partnerships
i) Operationalizing the Global Compact on Refugees: As a follow-up to the Global Refugee Forum (GRF), the 
RB undertook a pledge analysis and identified 517 pledges (from the 1,403 total pledges made) which are potentially 
relevant to the 11 EHAGL countries. In line with the GRF follow-up roadmap prepared by HQ, the Bureau is supporting 
country operations in GRF implementation with technical support and guidance through virtual meetings and tools. In 2022, 
the roadmap will inform key activities of the Regional Bureau .
Despite COVID19 and other constraints, UNHCR teams in the region made meaningful progress on the pledge follow 
up. Government-led pledge follow-up consultations have taken place in a number of countries, including Somalia and 
Uganda. In a number of EHAGL countries, including Rwanda and Somalia, pledge implementation strategies, roadmaps 
and action plans have been prepared. Despite the COVID-19 pandemic, tangible results have been achieved in Burundi, 
Ethiopia, Kenya, Rwanda, and Uganda. Progress has also been made in mobilizing resources supporting pledge 
matching. The World Bank, Germany and Denmark are some of the donors that made pledges matching the various 
government pledges on education, protection and livelihoods in Ethiopia, Kenya, Rwanda, and Uganda. 
Social protection and displacement financing are the key thematic focus areas of GCR operationalization in the Region. 
An internal study on the inclusion of Persons of Concern (PoC) to national COVID19 responses in Africa was undertaken 
with recommendations focusing on 3 countries (Djibouti, Kenya, Rwanda). Support to country operations in their 
advocacy for PoC inclusion with governments into WB-funded national systems are ongoing (Ethiopia, Somalia, South 
Sudan). The regional working group on Social protection and forced displacement will be revived in 2021 and led by the 
Senior Social Protection. Discussions with partners on displacement financing are taking place –  the Regional Durable 
Solutions Secretariat (ReDSS) is leading a regional study on displacement financing (case study focus on Ethiopia, 
Kenya, Somalia). UNHCR is also closely working with PSP and the innovative financing team to see how good examples 
from the region can be scaled-up or replicated.
In 2022, the comprehensive responses that will be pursued  include active and meaningful participation of development 
partners and the private sector, not only in financing the refugee response, but in including refugees and host 
communities in their programs by mainstreaming forced displacement in their analytical and operational activities as well 
as in their policy dialogue with governments. The EHAGL region has been pioneering on development responses to 
forced displacement and UNHCR’ s engagement with development actors. UNHCR has worked with the World Bank to 
publish two regional studies on forced displacement.
ii) Humanitarian and Development Nexus:In 2022, the Regional Bureau will maintain its key partnerships 
bilaterally and in the context of larger inter-agency humanitarian and development coordination processes and 
mechanisms, and will further enhances its collaboration with traditional partners through innovative approaches (such as 
with UNICEF, UNDP, WFP), develop new partnerships (such as with UN-Habitat, UNEP and UN Women) and engage 
strategically with development actors, financial institutions and a wider range of civil society actors. The partnerships at 
the regional level will be guided by UNHCR’ s overarching goals to promote the progressive implementation of the Global 
Compact on Refugees, to strengthen the protection of refugees, IDPs, returnees and stateless persons in the context of 
the COVID-19 response and other emergencies throughout the region and to contribute to the resilience of POCs and 
their host communities.
iii) Resource Mobilization: The Regional Bureau will further intensify its engagement with diverse stakeholders to 
strengthen resource mobilization and joint advocacy. This involves close collaboration with other UN agencies, 
development actors, governments, civil society organizations and private sector partners, among others. UNHCR will 
continue to exercise its leadership role in line with the Refugee Coordination Model, and in particular through the RRRPs 
for the Ethiopia and the South Sudan situations. The implementation of the Joint Refugee Return and Reintegration Plan 
(JRRRP) for Burundi refugees co-led with UNDP will be crucial to support the resilience and sustainable reintegration of 
increasing numbers of Burundian refugee returnees. The JRRRP recognizes the need to engage at an early stage with 
development partners, as well as a broader array of stakeholders to ensure the sustainability of returns and stabilization 
of affected communities.
iv) Sustainable Development Goals and UN Reform: UNHCR’ s Regional Bureau will strengthen its 
engagement with partners in the context of the UNDS reform and the promotion of the Sustainable Development Goals 
(SDGs). In the spirit of “ leaving no one behind”  and to bring coherence to UNHCR’ s approach to the 2030 Agenda, 
UNHCR will strengthen protection and pursue solutions for populations of concern through innovative approaches and 
partnerships which are key elements of UNHCR’ s overall strategic engagement in support of the SDGs. UNHCR will 
pursue strengthened collaboration with UNDP in the framework of the UNSDG process, and development, peace, and 
resilience projects throughout the region. The objective of inclusion is to reduce the number of populations of concern 
and host communities impacted by marginalization, lack of access to rights and poverty through:
*	Promoting an enabling legal framework and realizing their participation in national and local social and economic 
development;
*	Ensuring the inclusion of refugees and other persons of concern in national services, such as education, health, 
access to livelihoods and social services including social safety nets.
*	Building effective approaches to resilience and solutions that assist them and host governments to better 
manage and overcome the consequences and effects of displacement.
The objective of the new partnerships is to enlarge and strengthen the contributions of a broader range of stakeholders 
to effectively address the challenges to the achievement of the SDGs posed by forced displacement and statelessness, 
further exacerbated by the COVID-19 pandemic.
Addressing Mixed Movements. UNHCR’ s co-lead with IOM of the Opportunities/Issues Based Coalition (O/IBC) 7 
on Forced Displacement and Migration as part of the UNSDG process is an important opportunity to further strengthen the 
humanitarian-development nexus. The inclusion of forced displacement among the key themes for the regional development 
agenda reflects the growing recognition by diverse UN actors of the interlinkages between forced displacement issues and 
the development and stability of countries in Africa. The O/IBC on Forced Displacement and Migration supports the 
implementation of the Global Compact on Refugees and the Global Compact on Migration at a regional level and aims at 
promoting the SDGs based on the principle of “ leaving no one behind” . With 16 UN agencies participating in this O/IBCs, the 
forum provides an opportunity for UNHCR to develop strengthened synergies with UN agencies and include issues related to 
forced displacement strategically in the regional development agenda.
v) Self-Reliance through Inclusion:  A key priority of UNHCR’ s strategic engagement with partners in 2022 will be 
the inclusion of POCs in self-reliance projects across the humanitarian-development nexus.  The main coordination forum is 
the UNHCR-led Regional Economic Inclusion Working Group that involves a wide range of actors and meets on a 
monthly basis. In the context of this inter-agency coordination mechanism, partners will contribute to:
1.	Knowledge-sharing.
2.	Implementation of GRF pledges (In EHAGL a total of 38 pledges target Jobs and Livelihoods).
3.	 employment in private sector.
4.	Inclusion and fundraising for self-reliance programmes by NGOs; with a particular focus on facilitating the 
implementation of the Poverty Alleviation Coalition (PAC) at country level and at regional level in cooperation 
with ReDSS;
5.	 Strengthen the MADE51 artisanal regional network and increase access of refugees to global value chains.
6.	Regional implementation of global MOUs with other UN agencies, such as ILO, FAO, WFP, UN-Habitat and 
UNDP;
7.	Identification of new partners and opportunities in the digital economy, green jobs and the care economy. 
8.	Financial inclusion through workshops with financial service providers and regulators.
9.	Engagement with DFIs and donors to consider providing market-conducive incentives/blended finance for FSPs 
and investors during and in the aftermath of the crisis.1.2 - Planning scenario
Overarching Planning Assumption: The interim strategy assumes that the East, Horn and Great Lakes remains 
one of the most politically dynamic regions on the continent. The region remains vulnerable to disruptions related to a 
changing climate, with alternating drought/flood cycles, resulting in food insecurity, loss of livelihoods and forced 
displacement. High population growth rates, low employment opportunities and the socio-economic impact of the 
pandemic put real pressure on governments to deliver gains in human development and to create opportunities for their 
citizens as well as refugee populations.
Planning Scenario1.	A growing trend of reduction in asylum space in the region call for the Bureau’ s planning scenario to prioritize 
support to effective reception, registration and fair and efficient refugee status determination procedures, as well 
as promoting protection and solution frameworks working in partnership with Regional Economic Communities 
(RECs), the ICGLR, the AU and other stakeholders.  
2.	Ensuring effective inter-agency emergency responses to multiple and multifaceted refugee and IDP crises will be 
of the utmost priority in 2022. The Bureau will underpin its programming through in-depth conflict analysis to 
ensure predictable refugee, IDP and returnee responses.
3.	IDPs account for over third of the persons of concern in the region and will therefore be a priority within the 
planning scenario. In 2022, UNHCR will reinforce its strategic leadership through enhanced and predictable 
clusters for Protection, Camp Coordination and Camp Management (CCCM), and Shelter.
4.	Protection and expansion of humanitarian access to persons of concern will be critical in 2022 in all key 
operations. The insecurity in conflict affected countries has led to the denial of access to refugees in Ethiopia, 
Tanzania and South Sudan and Sudan. Restoring access to refugees and measures to preserve the 
humanitarian and civilian character of camps will be promoted.
5.	The COVID-19 pandemic is having a significant impact on the health and economic status of IDPs, refugees and 
hosting communities therefore ensuring refugees IDPs are integrated in national development plans, UN 
Sustainable Development Cooperation Frameworks. The GCR indicators will also need to mainstream with 
Governments and the UN system.
6.	The Bureau’ s Planning Scenario for 2022 will seek to broaden strategic engagements, expand its donor base 
and advance prospects for solutions.
 3 - Vision and Strategic Direction
No narrative found.
1.4 - Strategic focus and prioritization
In 2022, the Bureau will focus on the following strategic priority areas: ensuring a rapid, flexible and coordinated 
humanitarian response to emergency and protracted refugee and IDP situations in the region; supporting progress on 
regional commitments towards the eradication of statelessness, supporting progress on multi-stakeholder commitments 
to facilitate the inclusion and/or integration of refugees in alignment with global frameworks and strategic directions; and 
leveraging opportunities for comprehensive and coordinated regional approaches to solutions for protracted 
displacement in and from Burundi, South Sudan, Sudan, Somalia and the Democratic Republic of Congo.
To support efficient and effective delivery of protection and solutions in the region, within the mid-term goals outlined 
under the Bureau’ s vision and strategic direction, the following key areas of focus have been identified:
1.	The Bureau will engage in robust advocacy to support operational efforts to preserve asylum space, build 
national asylum capacity, maintain active tri-cluster leadership and delivery in priority IDP situations, eradicate 
statelessness, progress on all forms of inclusion and realise durable solutions through strategic advocacy at 
global, continental and regional level, with a focus on leveraging regional agency representations, RECs, MDBs, 
IFIs, Special Envoys and others, and ensuring enhanced cross-Bureau coordination where relevant and 
required.
2.	The Bureau will, to the extent possible, ensure sufficient financial and human resources are 
allocated to country operations to facilitate alignment with corporate-level objectives and initiatives and 
manage resources and assets within the region in line with the authorities and conditions set forth in the 
Administrative Instructions on Resource Allocation Framework. Resource allocation will be guided by regional 
and global strategic priorities.  
3.	The Bureau will act as a collaborative information hub that facilitates knowledge exchange, 
and enables evidence-based protection and solutions interventions.  In particular, the Bureau 
will focus on effective and comprehensive utilization of proGres v4 and digital case management systems, 
strengthening and harmonizing protection information management systems and tools, data protection oversight, 
improving the collection, analysis and dissemination of socio-economic data, providing sound political analysis to 
inform emergency and preparedness response and planning for solutions, and supporting the identification, 
dissemination and replication of lessons learnt across operations in the region.
4.	The Bureau will improve joint learning and enhance regional capacity on protection and 
solutions with an added focus on evidence-informed operational planning, management of risk and 
preservation of integrity through the contextualization of global policy and guidelines, support to learning and 
development, emergency or surge capacity, quality assurance, and the provision of timely and comprehensive 
technical guidance in the areas of CBI, child protection, GBV, health, education, livelihoods, financial inclusion, 
nutrition, WASH, monitoring and evaluation, and risk and integrity. 
5.	The Bureau will maintain and expand stakeholder engagement in support of protection and 
solutions for displacement-affected communities in the region, with a focus on the active leadership of an 
Inter-Agency Regional Strategic Engagement Group on Protection and Solutions, continued operationalization of 
the CRRF, implementation of HLS/GRF pledges as well as regional commitments under the Kampala and 
Djibouti Declarations, strengthened partnerships with development, peacebuilding actors, and the private sector, 
and strategic engagement with regional actors.
6.	The Bureau will continue to  develop and enhance internal control activities drawing upon 
operational context, critical and important facts, findings and recommendations from IOIS and IGO, lessons 
learned experiences, and periodic reviews of operational and partner risk registers and other pertaining 
information including Internal Control Questionnaires (ICQs). Recommendations from external oversight bodies 
and partner audits will be assessed and incorporated into second line of defense mechanisms where applicable.

1.5 - Assumptions and risk management

Assumption: 
The interim strategy assumes that the East, Horn and Great Lakes remains one of the most politically dynamic regions 
on the continent. The region remains vulnerable to disruptions related to a changing climate, with alternating 
drought/flood cycles, resulting in food insecurity, loss of livelihoods and forced displacement. High population growth 
rates, low employment opportunities and the socio-economic impact of the pandemic put real pressure on governments 
to deliver gains in human development and to create opportunities for their citizens as well as refugee populations.
Risk Management:
Key strategic risks include fragile security and political situations, risks associated with asylum (access, quality, the 
civilian & humanitarian character), funding constraints, food insecurity, worsening socio-economic opportunities due to 
COVID 19 and the adverse effects of climate change.
Key opportunities are the GCR and integration of POCs into national systems, the Solutions strategy for Sudan & South 
Sudan and UNHCR’ s engagement with development finance institutions and the private sector. 
Important mitigation measures include effective emergency preparedness and response, monitoring and advocacy on 
asylum, the use of a multi stakeholder approach for fundraising, livelihoods/self-reliance interventions.
ive emergency preparedness and response, monitoring and advocacy on asylum, the use of a multi stakeholder 
approach for fundraising, livelihoods/self-reliance interventions.
 










1.6a - Operational resources and capacity

No narrative found.









1.6b - Operational resources, capacity and presence

No narrative found.



1.7 - Security management

In 2022, the security environment in the 11 countries of the East, Horn and Great Lakes region of 
Africa is not expected to experience significant changes. Terrorism, armed conflict, civil unrest and 
criminality will remain a concern in the region. The situation varies from one country to the other and 
contributes to the risks of instability, impacting the safety of local population, operations and 
humanitarian personnel.
 
Kenya is expected to hold general elections on 9 August 2022, and South Sudan’ s elections are 
also scheduled for 2022.  The last two general elections in Kenyan turned violent, resulting in 
casualties. The security environment in Sudan, South Sudan, Somalia and Ethiopia are expected to 
continue to be high-risk environments, presenting hardships and challenges for UNHCR operations.
 
Tanzania has been impacted by acts of terrorism along its border with Mozambique, and Uganda 
has been impacted by the insecurity in the neighboring Eastern DRC and South Sudan. 
 
Vectors of insecurity across the region relate to: 
*	Political and socio-economic grievances within countries,
*	Continued armed conflict linked either to terrorism or ethnic/tribal conflict,
*	High level of criminality,
*	Countries’  limited capacity to manage conflict and crisis at its borders,
*	Poverty and Food insecurity,
*	Impact of COVID-19,
*	Impact of climate changes (drought/floods) limited access to resources
These drivers of conflict have resulted in violence in the past, impacting populations, humanitarian 
workers and particularly UN and partners’  national staff. Field operations with the support of security 
officers will maintain their presence to deliver humanitarian assistance. Staff will be trained and 
sensitized on security awareness within the UN’ s framework of ‘ Duty of Care’ . 
 
To support staff and operations in 2022, the Security Unit will continue to create security awareness 
and to ensure full compliance with security measures and mandatory security training including: 
BSAFE, SMT training and SSAFE training. 
 
COVID-19 will continue to require precautions, the security unit will maximize online programs to 
create awareness on COVID-19 measures. 
 
Currently, field operations are supported by 77 security officers including 16 International staff and 
coordinated by the Regional Field Security Coordination based at the EHAGL Regional Bureau. 
The role of the Regional Security Coordinator entails: 
-       Providing practical and professional security advice to the Regional Director in managing safety of 
staff, partners, operations, and persons of concern. 
-       Monitoring situations and events in the region and gathering, analyze information to identify 
emerging trends and providing early warning of any potential threats that may affect the safety and 
security of UNHCR personnel, premises, assets and impact programme delivery. 
-       Conducting regular missions to countries in the region, providing security advice, training, 
Security Risk Assessments, security risk management measures, operational plans and criticality 
assessments, 
-       Overseeing beneficiary security issues and security-related administrative issues. 
Plan for 2022
1.	Security oversight and review (SOAR): This initiative started in 2021 by FSS which will 
continue in 2022. The regional security unit will engage the field where specific threats are of 
concern. The dialogue should lead to determining the required support to the field and 
eventually to the escalation of issues to the Security Steering Committee.
2.	For better monitoring and supporting countries in the region, the concept of 'Operations 
Room' will be established and operationalized at the bureau level. This will help to closely 
follow up security incidents, analyze security trends and issue security alerts. 
3.	Engage a security officer/analyst to support the 'Operational Room' of the regional security 
coordination. 
 
 










1.8 - Resource mobilization strategy

The Resource base for the region is expected to remain relatively stable in 2022. Funding from government donors is 
expected to be maintained at relatively similar levels to past years with possible increases to support emergencies and 
emerging situations in the region. Additional efforts will however be made in accessing pooled funds and engaging with 
emerging donors and the private sector to increase funding in the region.
 
UNHCR’ s Global Resource mobilization goals aim to secure sufficient, predictable, flexible and multi-year funding for 
UNHCR operations; Sustain and protect top and major donors,  diversify the donor base accessing new funding streams, 
and support growth in public donors  to keep pace with needs; Demonstrate UNHCR’ s value for money, effectiveness 
and speak to the impact of our work and Ensure transparency and accountability to donors and affected populations.
 
The Bureau will continue to support country operations in the region in their Resources Mobilization efforts, engaging 
donors in the region, highlighting ongoing and evolving needs and strengthening donor engagement by keeping donors 
updated on key operational needs. In coordination with the HQ, the Bureau will explore funding opportunities from 
donors, including in furthering the objectives of the GCR, to bridge the Humanitarian and Development nexus.
 
The Bureau will ensure that donors are kept abreast on the operational issues in the region through regular briefings and 
sharing operational updates and organizing individual or thematic briefs as may be requested or deemed necessary. 
Donor missions will also be organized when feasible to bring donors to appreciate needs of PoCs as well as the impact 
of their funding.
 
The Bureau will work to increase communication that conveys the effectiveness of UNHCR and partners work and its 
impact on people of concern, and by extension the value of supporting that work. These communications will be data 
driven and demonstrate the human impact of donor support, further enhancing donor visibility.
 
The Bureau will seek to boost collaboration with partners to jointly mobilize resources for people of concern through 
RRRPs and other joint fundraising platforms and donor outreach.
 
Work to diversify the donor base and funding streams including strengthened access to pooled funds (CBF, CERF, PBF) 
and other specific government funding streams for development activities and other thematic funds that may exist either 
at embassy/country levels or other will continue. The Bureau will follow-up with donor countries on GCR pledges and, in 
coordination with the PSP team, identify and support specific countries with Private sector engagement.
 










2 - Results Chains

2.1 - Regional Bureau EHGL ABC

2.1.1 - Impact Statement 
Persons of concern enjoy legal and social protection to realize their rights (IA1: Protect)



2.1.1.7 - Outcome Statement 
Country operations are effectively monitoring and advocating for Government procedures 
permitting safe access to territory and indiscriminate access to protection-sensitive asylum 
procedures including registration and profiling, legal identity and documentation. (OA1: 
Access/Doc)


2.1.1.5.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.1.5.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.1.5.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.1.7.5 - Output Statement (11H) 
Capacity building, guidance and support to country operations and national authorities 
regarding PoC access to asylum systems including registration and procedures provided.

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.1.7.6 - Output Statement (11J) 
Guidance and support to country operations in implementation of DIP SOPs for addressing 
refoulement or imminent risk of refoulement, provided

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.1.4 - Outcome Statement 
Country operations in EHAGL are effectively supporting Governments in establishing and 
implementing quality Refugee Status Determination procedures that are fair, efficient, 
adaptable and that operate with integrity, including in mixed flow situations, and/or that 
support Government in putting in place Mandate RSD procedures that adhere to, at least, to 
the 25 August 2020 Administrative Instructio (OA2: Status)


2.1.1.2.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.1.2.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.1.2.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.1.4.5 - Output Statement (10O) 
Advice, guidance and support (including training and individual casework support) provided 
to UNHCR country operations, States and other partners/stakeholders to ensure quality 
asylum decision-making and the creation and strengthening of national asylum/RSD systems

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.1.4.6 - Output Statement (10Q) 
Monitoring and oversight of operations' coherent and consistent implementation of guidance 
and policies including UNHCR's RSD procedural standards

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.1.5 - Outcome Statement 
Country operations have effectively been advocating for relevant international and regional 
instruments on refugee and IDP protection and eradication of statelessness to be adopted by 
the AU and/or for countries under RBEHAGL's remit to accede or ratify regional instruments 
or lift reservations in relation to the same (OA3: Policy/Law)


2.1.1.3.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.1.3.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.1.3.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.1.5.5 - Output Statement (11C) 
All country operations have received strategic advocacy support for adoption, accession to, 
ratification and/or lifting of reservations of relevant international and regional treaties on 
refugee and IDP protection, and eradication of statelessness.

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.1.5.6 - Output Statement (11D) 
Advocacy and stakeholders mobilization is pursued effectively with ICGLR towards adoption 
of the draft Protocol to APHRC on eradication of statelessness in Africa

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.1.6 - Outcome Statement 
Country operations have effectively been advocating for adequate and comprehensive 
domestic laws reforms to align the protection of refugees, IDPs and stateless as well as 
prevention and reduction of statelessness with international and regional standards (OA3: 
Policy/Law)


2.1.1.4.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.1.4.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.1.4.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.1.6.5 - Output Statement (11E) 
All country operations have received the technical guidance/support to align domestic 
legislation with international and regional standards on refugee and IDP protection, and 
prevention and reduction of statelessness

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.1.6.6 - Output Statement (11G) 
Guidance and support provided to the ICGLR Secretariat for the development and adoption of 
a regional policy and programmatic framework on civil registration, including birth 
registration for the prevention of statelessness (by 2022)

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.1.6.7 - Output Statement (12D) 
All country operations have received support/tailored guidance to ensure that their domestic 
legal, policy and administrative frameworks is aligned with the principle of universal birth 
registration for the prevention and reduction of statelessness and the relevant international, 
regional and ICGLR related policies and good practices promoted in the ICGLR region

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".
















2.1.2 - Impact Statement 
Persons of concern receive support and assistance to realize their rights in safe environment 
(IA2: Assist)



2.1.2.9 - Outcome Statement 
Country operations in EHAGL are designing and delivering emergency and protracted 
assistance to refugees and IDPs that effectively incorporate UNHCR's standards for SGBV 
(OA4: GBV)


2.1.2.7.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.2.7.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.2.7.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.2.9.5 - Output Statement (11O) 
Guidance, tools and technical support provided to country operations for advocating to target 
persons of concerns, partners and host communities to meet GBV standards and milestones

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.2.10 - Outcome Statement 
Country operations in EHAGL have effectively contributed to and advocated for safe access of 
POCs to child protection, risk mitigation and response services and programmes (OA5: 
Children)


2.1.2.8.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.2.8.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.2.8.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.2.10.5 - Output Statement (11Q) 
Operations have received technical support and guidance from the regional Bureau to meet 
child protection standards

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.2.10.6 - Output Statement (11R) 
Regional strategy for the implementation of Child Protection Minimum Standards developed

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.2.7 - Outcome Statement 
Country operations in EHAGL have effectively contributed to and advocated for POCs and 
host communities' access to sufficient basic and domestic items including improved food 
security (OA8: Well-being)


2.1.2.5.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.2.5.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.2.5.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.2.7.5 - Output Statement (10Z) 
Advocacy, technical support, and guidance provided to operations for provision of food 
assistance (in-kind and cash) core relief items including energy for cooking to refugees.

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.2.7.6 - Output Statement (110) 
Joint assessments, plans and strategies agreed with WFP to address malnutrition and food 
security needs of refugees.

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.2.8 - Outcome Statement 
Country offices in the region effectively established, improved and maintained shelter and 
infrastructure (OA9: Housing)


2.1.2.6.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.2.6.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.2.6.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.2.8.5 - Output Statement (111) 
Capacity development support provided to UNHCR operations and partner staff on innovative 
shelter and infrastructure design and development to ensure PoC’s access equitable and 
adequate emergency, transitional and permanent housing as well as sustainable settlement 
planning

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.2.4 - Outcome Statement 
Country operations in EHAGL have effectively contributed to and advocated for access of 
POCs to comprehensive quality health care at the same level as nationals (OA10: Health)


2.1.2.2.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.2.2.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.2.2.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.2.4.5 - Output Statement (10U) 
Technical support, guidance and tools including recruitment of SENS consultations, reviewing 
TORs, review of analysis of nutritional status provided to operations in EHAGL region for 
conducting assessment and analysis of nutritional well-being of refugees.

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.2.4.6 - Output Statement (11T) 
Guidelines, capacity building, advocacy and technical support provided to operations on 
innovative health solutions, emergency response and inclusion of PoCs into national health 
systems

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.2.5 - Outcome Statement 
Country operations in EHAGL have effectively contributed to and advocated for increased 
access of POCs to inclusive, equitable and quality education (OA11: Education)


2.1.2.3.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.2.3.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.2.3.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.2.5.5 - Output Statement (10V) 
Capacity building, advocacy and technical support provided to operations on equitable, 
inclusive & quality education programming and on inclusion of refugees in national education 
systems

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.2.6 - Outcome Statement 
Country operations in EHAGL have effectively contributed to and advocated for a POCs and 
host communities' access to and use of water, sanitation and hygiene facilities that comply 
with UNHCR and/or national minimum WASH standards (OA12: WASH)


2.1.2.4.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.2.4.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.2.4.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.2.6.5 - Output Statement (10X) 
Review of existing operations, strategies, polices, guidelines for capacity development, 
advocacy and technical support to UNHCR and partners staff to improve satisfactory 
conditions of WASH completed jointly by the RB and HQ

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".
















2.1.3 - Impact Statement 
Persons of concern thrive as part of empowered community (IA3: Empower)



2.1.3.5 - Outcome Statement 
Country operations in EHAGL effectively integrate UNHCR's policy on and operational 
standard operational procedures for Age, Gender and Diversity when designing and delivering 
emergency and protracted assistance, and protection and solutions and when advocating for 
inclusion to POC in national systems (OA7: Community)


2.1.3.3.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.3.3.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.3.3.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.3.5.5 - Output Statement (11W) 
Refugee-led organizations have been supported through a regional initiative

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.3.5.6 - Output Statement (11X) 
Operations have received RB’s support in deploying digital portals to enable POCs to access 
their personal data and make online requests for services.

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.3.5.7 - Output Statement (11Y) 
Country operations are supported on engagement of affected people in development of 
UNHCR strategies and policies including operationalization of AAP

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.3.4 - Outcome Statement 
Country operations in EHAGL have effectively supported and advocated for enhancement of 
livelihood opportunities, economic inclusion, self-reliance and inclusion in government social 
protection systems and programmes. (OA13 Livelihood)


2.1.3.2.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.3.2.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.3.2.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.3.4.5 - Output Statement (113) 
Guidance and support provided to operations for innovative and market-based livelihoods 
programming in line with development priorities of the host country.

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.3.4.6 - Output Statement (114) 
Market and evidence-based livelihoods programming for POCs is enabled in the region 
through the creation of a solid evidence based information dissemination as shared through 
the economic Inclusion working group “ Economic Inclusion East Africa".

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.3.4.7 - Output Statement (12F) 
Understanding and evidence for incluson of PoC in social protection is enhanced.

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".
















2.1.4 - Impact Statement 
Persons of concern enjoy durable and alternative pathways to solutions (IA4: Solve)



2.1.4.6 - Outcome Statement 
Country operations in the region effectively supporting/collaborating with the governments 
to ensure safe, dignified and voluntary returns for refugee and internally displaced 
populations (OA14: Return)


2.1.4.4.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.4.4.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.4.4.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.4.6.5 - Output Statement (11A) 
Guidance and support provided to countries in developing country strategies, tools and 
methodologies to support evidence-based planning for returns and reintegration

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.4.6.6 - Output Statement (11B) 
Regional dialogue and frameworks for sustainable returns and reintegration supported

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.4.4 - Outcome Statement 
Country operations in the region are effectively identifying and referring refugees meeting 
criteria for solutions through resettlement and complementary pathways (OA15: Resettle)


2.1.4.2.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.4.2.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.4.2.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.4.4.5 - Output Statement (115) 
Guidance and support to country operations on development of complementary pathways 
programs and resettlement planning and submissions provided

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.4.5 - Outcome Statement 
Country Operations have effectively advocated for greater reduction of statelessness through 
local integration and other legal solutions that enable access to proof of legal identity for 
stateless persons and those with undetermined nationality (OA16: Integrate)


2.1.4.3.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.4.3.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.4.3.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.4.5.5 - Output Statement (118) 
Guidance and support provided to countries in developing and implementing policies and 
strategies in support of local solutions

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.4.5.6 - Output Statement (119) 
Guidance and support provided to ICGLR for the development of a comprehensive strategy 
on durable solutions for refugees in the Great Lake

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".
















2.1.5 - Impact Statement 
Persons concern and partners receive effective and efficient support from UNHCR (IA5: 
HQ/RB Support)



2.1.5.4 - Outcome Statement 
Country operations in the region implement robust systems and processes for evidence 
based programme planning and implementation including internal controls, risk 
management, monitoring, evaluation and reporting in accordance with UNHCR's policies and 
guidelines (EA17: Systems)


2.1.5.2.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.5.2.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.5.2.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.5.4.5 - Output Statement (105) 
Support provided to operations to set up effective implementation and monitoring 
arrangements

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.5.4.6 - Output Statement (106) 
Technical guidance provided to support country operations to achieve implementation and 
scaling up of CBI across the sectors

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.5.4.7 - Output Statement (107) 
Community of Practitioners (COP) fostered, and exchanges facilitated in response to demand. 
Evaluation findings from UNHCR evaluations and evaluations from other international 
organizations shared with regional stakeholders

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.5.4.8 - Output Statement (108) 
Country office and RB staff in EHAGL planning and/or implementing evaluations have 
received evaluation technical assistance, quality assurance and process mentoring

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.5.4.9 - Output Statement (109) 
Enhancement of internal control/oversight dashboard and formal communications with 
Country Office representatives and Admin/Finance, project control staff

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.5.4.10 - Output Statement (10A) 
Effective support, review and approval process related to Interim and Multi-year Strategies in 
place

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.5.4.11 - Output Statement (120) 
Oversight and technical assistance provided to country operations and the regional bureau on 
identification and mitigation of key and emerging risks

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.5.4.12 - Output Statement (12B) 
Regional easy-to-access M&E surge capacity established

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.5.4.13 - Output Statement (12E) 
Support provided to operations to set up effective implementation and monitoring of 
financial, budgetary and oversight activities

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.5.10 - Outcome Statement 
Country operations have strengthened and harmonized protection information management 
systems and tools, data protection oversight, and collection, analysis and dissemination of 
socio-economic data to provide sound political analysis that can inform emergency and 
preparedness response and planning for solutions (EA17: Systems)


2.1.5.8.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.5.8.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.5.8.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.5.10.5 - Output Statement (11N) 
Country operations and the RB receive technical assistance from RB DIMA on data collection, 
storage, cleaning, analysis and dissemination for operational data activities

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.5.11 - Outcome Statement 
Country operations in EHAGL effectively implement UNHCR's Policy on and Operational 
Guidelines on addressing Fraud Committed by Persons of Concern to safeguard the access of 
persons of concern to protection, assistance and durable solutions (EA17: Systems)


2.1.5.9.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.5.9.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.5.9.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.5.11.5 - Output Statement (10M) 
Country operations have received guidance and technical support to implement UNHCR’s 
Policy on and Operational Guidelines on addressing Fraud Committed by Persons of Concern, 
etc. and strengthening the integrity of protection, assistance and solution procedures

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.5.9 - Outcome Statement 
Operational Support and Supply Chain are effective and efficient to meet operational needs 
including emergency preparedness and response (EA18: Support)


2.1.5.7.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.5.7.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.5.7.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.5.9.5 - Output Statement (10B) 
Guidance, support and quality assurance related to finance, administration, supply and 
programme management provided

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.5.9.6 - Output Statement (10C) 
Regional Committee on Contracts and Asset Management Board works effectively and 
supports operational needs

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.5.9.7 - Output Statement (10D) 
Security management in the Region reinforced

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.5.9.8 - Output Statement (11Z) 
Coordinate, provide guidance and feedback on drafting of Contingency Plans and other 
concerted actions

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.5.9.9 - Output Statement (123) 
Real-time and continuous risk analysis conducted

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.5.12 - Outcome Statement 
Country operations in the region and the Regional Bureau are strengthened for effective and 
efficient IT service delivery to meet the operational needs including the Preparedness and 
Response capabilities (EA18: Support)


2.1.5.10.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.5.10.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.5.10.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.5.12.5 - Output Statement (124) 
Mandatory and other capacity building trainings/orientations are conducted for UNHCR and 
partner IT workforce

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.5.12.6 - Output Statement (127) 
Guidance and support provided for the operations in the region to ensure smooth adoption 
of policies and effective use of, IT standards and procedures, refugees system and related 
applications

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.5.12.7 - Output Statement (128) 
Country Operations developed, tested and implemented the IT preparedness, contingency 
and response plans in compliance with the Emergency Policy

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.5.5 - Outcome Statement 
Country Operations in the region have reduced incidents of fraud, harassment and SEA and 
they are responding pro-actively to incidents as per UNHCR's Guidelines, policies and 
standards (EA19: People)


2.1.5.3.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.5.3.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.5.3.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.5.5.5 - Output Statement (103) 
Mandatory and other capacity building trainings are conducted for UNHCR and partner staff

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.5.5.6 - Output Statement (104) 
Training provided on human resources, finance, administration, supply and programme 
management, etc. based on training needs assessments

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.5.6 - Outcome Statement 
Relevant regional bodies, Governments in the EHAGL region, the private sector, international 
partners and diverse civil society actors have been effectively engaged in support of 
protection and solutions for displacement-affected communities (in the region), through a 
new UNHCR-led Inter-Agency Regional Strategic Engagement Group and UNHCR's active 
participation in other regional inter-agency coor (EA20: External)


2.1.5.4.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.5.4.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.5.4.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.5.6.5 - Output Statement (10E) 
Engagement in regional platforms and coordination structures is ensured and interagency 
processes in country operations are supported

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.5.6.6 - Output Statement (10F) 
Partnerships with UN, NGOs, MDBs, IFIs and other development partners and other actors 
are strengthened and support to country operations is provided

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.5.6.7 - Output Statement (10G) 
Operations and HQ planning and processes related to the Global Refugee Forum have 
supported regionally to optimize realization of GCR and GRF objectives.

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.5.6.8 - Output Statement (10H) 
Regional engagement of partners in a whole-of-society approach supports GRF development 
and pledge implementation.

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.5.7 - Outcome Statement 
Communications and media are effectively used at country and at RB levels to build 
awareness of, empathy with, and concrete support for refugees and other forcibly displaced 
persons in the East, Horn and Great Lakes Region in line with the bureau’s strategic objectives 
and to position UNHCR as an effective, impactful organization (EA20: External)


2.1.5.5.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.5.5.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.5.5.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.5.7.5 - Output Statement (10J) 
Media engagement and external communications are strengthened through support to 
country operations, effective coordination with HQs and the three regional bureaus in Africa, 
and production of quality external products reflecting prioritized advocacy aims and resource 
mobilization efforts.

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".

















2.1.5.8 - Outcome Statement 
Donors are engaged and kept informed of operational needs in the EHAGL region to mobilize 
resources and sustain support to refugees and persons of concern in the region (EA20: 
External)


2.1.5.6.1 - Expected outputs

No narrative found in section "05_RF_OtcS" with name "Expected outputs".




2.1.5.6.2 - Intended results

No narrative found in section "05_RF_OtcS" with name "Intended results".




2.1.5.6.3 - Assumptions and risk management

No narrative found in section "05_RF_OtcS" with name "Assumptions and risk management".



2.1.5.8.5 - Output Statement (10K) 
Regional inputs to Global Appeals, Global Reports, Supplementary Appeals, Donor 
submissions, Concept Notes and Donor reports are delivered

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".



2.1.5.8.6 - Output Statement (10L) 
Resource mobilization training and capacity development needs of the operations identified 
and addressed

Key Activities

No narrative found in section "06_RF_OtpS" with name "Key activities".
























Section 3: Resources

3.1 - Budget Overview
Current "Implementation - Draft" OP/OL budget as currently recorded in COMPASS



2022
Regional Bureau EHGL ABC
18,334,896.71
Kenya
18,334,896.71
Persons of concern enjoy legal and social protection to realize their rights (IA1: Protect)
6,239,711.60
Country operations are effectively monitoring and advocating for Government procedures 
permitting safe access to territory and indiscriminate access to protection-sensitive asylum 
procedures including registration and profiling, legal identity and documentation. (OA1: Access/Doc)
1,386,602.58
Capacity building, guidance and support to country operations and national authorities 
regarding PoC access to asylum systems including registration and procedures provided.
693,301.29
Guidance and support to country operations in implementation of DIP SOPs for addressing 
refoulement or imminent risk of refoulement, provided
693,301.29
Country operations in EHAGL are effectively supporting Governments in establishing and 
implementing quality Refugee Status Determination procedures that are fair, efficient, adaptable 
and that operate with integrity, including in mixed flow situations, and/or that support Government 
in putting in place Mandate RSD procedures that adhere to, at least, to the 25 August 2020 
Administrative Instructio (OA2: Status)
1,386,602.58
Monitoring and oversight of operations' coherent and consistent implementation of guidance 
and policies including UNHCR's RSD procedural standards
693,301.29
Advice, guidance and support (including training and individual casework support) provided to 
UNHCR country operations, States and other partners/stakeholders to ensure quality asylum 
decision-making and the creation and strengthening of national asylum/RSD systems
693,301.29
Country operations have effectively been advocating for adequate and comprehensive domestic laws 
reforms to align the protection of refugees, IDPs and stateless as well as prevention and reduction of 
statelessness with international and regional standards (OA3: Policy/Law)
2,079,903.87
Guidance and support provided to the ICGLR Secretariat for the development and adoption of a 
regional policy and programmatic framework on civil registration, including birth registration for 
the prevention of statelessness (by 2022)
693,301.29
All country operations have received the technical guidance/support to align domestic 
legislation with international and regional standards on refugee and IDP protection, and 
prevention and reduction of statelessness
693,301.29
All country operations have received support/tailored guidance to ensure that their domestic 
legal, policy and administrative frameworks is aligned with the principle of universal birth 
registration for the prevention and reduction of statelessness and the relevant international, 
regional and ICGLR related policies and good practices promoted in the ICGLR region
693,301.29
Country operations have effectively been advocating for relevant international and regional 
instruments on refugee and IDP protection and eradication of statelessness to be adopted by the AU 
and/or for countries under RBEHAGL's remit to accede or ratify regional instruments or lift 
reservations in relation to the same (OA3: Policy/Law)
1,386,602.58
All country operations have received strategic advocacy support for adoption, accession to, 
ratification and/or lifting of reservations of relevant international and regional treaties on 
refugee and IDP protection, and eradication of statelessness.
693,301.29
Advocacy and stakeholders mobilization is pursued effectively with ICGLR towards adoption of 
the draft Protocol to APHRC on eradication of statelessness in Africa
693,301.29
Persons of concern receive support and assistance to realize their rights in safe environment (IA2: Assist)
2,846,373.00
Country operations in EHAGL have effectively contributed to and advocated for access of POCs to 
comprehensive quality health care at the same level as nationals (OA10: Health)
569,274.60
Guidelines, capacity building, advocacy and technical support provided to operations on 
innovative health solutions, emergency response and inclusion of PoCs into national health 
systems
284,637.30
Technical support, guidance and tools including recruitment of SENS consultations, reviewing 
TORs, review of analysis of nutritional status provided to operations in EHAGL region for 
conducting assessment and analysis of nutritional well-being of refugees.
284,637.30
Country operations in EHAGL have effectively contributed to and advocated for increased access of 
POCs to inclusive, equitable and quality education (OA11: Education)
284,637.30
Capacity building, advocacy and technical support provided to operations on equitable, inclusive 
& quality education programming and on inclusion of refugees in national education systems
284,637.30
Country operations in EHAGL have effectively contributed to and advocated for a POCs and host 
communities' access to and use of water, sanitation and hygiene facilities that comply with UNHCR 
and/or national minimum WASH standards (OA12: WASH)
284,637.30
Review of existing operations, strategies, polices, guidelines for capacity development, advocacy 
and technical support to UNHCR and partners staff to improve satisfactory conditions of WASH 
completed jointly by the RB and HQ
284,637.30
Country operations in EHAGL are designing and delivering emergency and protracted assistance to 
refugees and IDPs that effectively incorporate UNHCR's standards for SGBV (OA4: GBV)
284,637.30
Guidance, tools and technical support provided to country operations for advocating to target 
persons of concerns, partners and host communities to meet GBV standards and milestones
284,637.30
Country operations in EHAGL have effectively contributed to and advocated for safe access of POCs 
to child protection, risk mitigation and response services and programmes (OA5: Children)
569,274.60
Regional strategy for the implementation of Child Protection Minimum Standards developed
284,637.30
Operations have received technical support and guidance from the regional Bureau to meet 
child protection standards
284,637.30
Country operations in EHAGL have effectively contributed to and advocated for POCs and host 
communities' access to sufficient basic and domestic items including improved food security (OA8: 
Well-being)
569,274.60
Advocacy, technical support, and guidance provided to operations for provision of food 
assistance (in-kind and cash) core relief items including energy for cooking to refugees.
284,637.30
Joint assessments, plans and strategies agreed with WFP to address malnutrition and food 
security needs of refugees.
284,637.30
Country offices in the region effectively established, improved and maintained shelter and 
infrastructure (OA9: Housing)
284,637.30
Capacity development support provided to UNHCR operations and partner staff on innovative 
shelter and infrastructure design and development to ensure PoC’s access equitable and 
adequate emergency, transitional and permanent housing as well as sustainable settlement 
planning
284,637.30
Persons of concern thrive as part of empowered community (IA3: Empower)
4,447,493.00
Country operations in EHAGL have effectively supported and advocated for enhancement of 
livelihood opportunities, economic inclusion, self-reliance and inclusion in government social 
protection systems and programmes. (OA13 Livelihood)
2,223,746.50
Market and evidence-based livelihoods programming for POCs is enabled in the region through 
the creation of a solid evidence based information dissemination as shared through the 
economic Inclusion working group “ Economic Inclusion East Africa".
741,248.83
Guidance and support provided to operations for innovative and market-based livelihoods 
programming in line with development priorities of the host country.
741,248.83
Understanding and evidence for incluson of PoC in social protection is enhanced.
741,248.83
Country operations in EHAGL effectively integrate UNHCR's policy on and operational standard 
operational procedures for Age, Gender and Diversity when designing and delivering emergency and 
protracted assistance, and protection and solutions and when advocating for inclusion to POC in 
national systems (OA7: Community)
2,223,746.50
Operations have received RB’s support in deploying digital portals to enable POCs to access 
their personal data and make online requests for services.
741,248.83
Refugee-led organizations have been supported through a regional initiative
741,248.83
Country operations are supported on engagement of affected people in development of UNHCR 
strategies and policies including operationalization of AAP
741,248.83
Persons of concern enjoy durable and alternative pathways to solutions (IA4: Solve)
3,509,273.11
Country operations in the region effectively supporting/collaborating with the governments to 
ensure safe, dignified and voluntary returns for refugee and internally displaced populations (OA14: 
Return)
1,403,709.24
Guidance and support provided to countries in developing country strategies, tools and 
methodologies to support evidence-based planning for returns and reintegration
701,854.62
Regional dialogue and frameworks for sustainable returns and reintegration supported
701,854.62
Country operations in the region are effectively identifying and referring refugees meeting criteria 
for solutions through resettlement and complementary pathways (OA15: Resettle)
701,854.62
Guidance and support to country operations on development of complementary pathways 
programs and resettlement planning and submissions provided
701,854.62
Country Operations have effectively advocated for greater reduction of statelessness through local 
integration and other legal solutions that enable access to proof of legal identity for stateless 
persons and those with undetermined nationality (OA16: Integrate)
1,403,709.24
Guidance and support provided to countries in developing and implementing policies and 
strategies in support of local solutions
701,854.62
Guidance and support provided to ICGLR for the development of a comprehensive strategy on 
durable solutions for refugees in the Great Lake
701,854.62
Persons concern and partners receive effective and efficient support from UNHCR (IA5: HQ/RB Support)
1,292,046.00
Country operations in the region implement robust systems and processes for evidence based 
programme planning and implementation including internal controls, risk management, monitoring, 
evaluation and reporting in accordance with UNHCR's policies and guidelines (EA17: Systems)
415,300.50
Support provided to operations to set up effective implementation and monitoring 
arrangements
46,144.50
Effective support, review and approval process related to Interim and Multi-year Strategies in 
place
46,144.50
Enhancement of internal control/oversight dashboard and formal communications with Country 
Office representatives and Admin/Finance, project control staff
46,144.50
Community of Practitioners (COP) fostered, and exchanges facilitated in response to demand. 
Evaluation findings from UNHCR evaluations and evaluations from other international 
organizations shared with regional stakeholders
46,144.50
Support provided to operations to set up effective implementation and monitoring of financial, 
budgetary and oversight activities
46,144.50
Technical guidance provided to support country operations to achieve implementation and 
scaling up of CBI across the sectors
46,144.50
Oversight and technical assistance provided to country operations and the regional bureau on 
identification and mitigation of key and emerging risks
46,144.50
Regional easy-to-access M&E surge capacity established
46,144.50
Country office and RB staff in EHAGL planning and/or implementing evaluations have received 
evaluation technical assistance, quality assurance and process mentoring
46,144.50
Country operations have strengthened and harmonized protection information management 
systems and tools, data protection oversight, and collection, analysis and dissemination of socio-
economic data to provide sound political analysis that can inform emergency and preparedness 
response and planning for solutions (EA17: Systems)
46,144.50
Country operations and the RB receive technical assistance from RB DIMA on data collection, 
storage, cleaning, analysis and dissemination for operational data activities
46,144.50
Country operations in EHAGL effectively implement UNHCR's Policy on and Operational Guidelines 
on addressing Fraud Committed by Persons of Concern to safeguard the access of persons of concern 
to protection, assistance and durable solutions (EA17: Systems)
46,144.50
Country operations have received guidance and technical support to implement UNHCR’s Policy 
on and Operational Guidelines on addressing Fraud Committed by Persons of Concern, etc. and 
strengthening the integrity of protection, assistance and solution procedures
46,144.50
Operational Support and Supply Chain are effective and efficient to meet operational needs including 
emergency preparedness and response (EA18: Support)
230,722.50
Coordinate, provide guidance and feedback on drafting of Contingency Plans and other 
concerted actions
46,144.50
Guidance, support and quality assurance related to finance, administration, supply and 
programme management provided
46,144.50
Regional Committee on Contracts and Asset Management Board works effectively and supports 
operational needs
46,144.50
Real-time and continuous risk analysis conducted
46,144.50
Security management in the Region reinforced
46,144.50
Country operations in the region and the Regional Bureau are strengthened for effective and 
efficient IT service delivery to meet the operational needs including the Preparedness and Response 
capabilities (EA18: Support)
138,433.50
Country Operations developed, tested and implemented the IT preparedness, contingency and 
response plans in compliance with the Emergency Policy
46,144.50
Mandatory and other capacity building trainings/orientations are conducted for UNHCR and 
partner IT workforce
46,144.50
Guidance and support provided for the operations in the region to ensure smooth adoption of 
policies and effective use of, IT standards and procedures, refugees system and related 
applications
46,144.50
Country Operations in the region have reduced incidents of fraud, harassment and SEA and they are 
responding pro-actively to incidents as per UNHCR's Guidelines, policies and standards (EA19: 
People)
92,289.00
Mandatory and other capacity building trainings are conducted for UNHCR and partner staff
46,144.50
Training provided on human resources, finance, administration, supply and programme 
management, etc. based on training needs assessments
46,144.50
Relevant regional bodies, Governments in the EHAGL region, the private sector, international 
partners and diverse civil society actors have been effectively engaged in support of protection and 
solutions for displacement-affected communities (in the region), through a new UNHCR-led Inter-
Agency Regional Strategic Engagement Group and UNHCR's active participation in other regional 
inter-agency coor (EA20: External)
184,578.00
Engagement in regional platforms and coordination structures is ensured and interagency 
processes in country operations are supported
46,144.50
Operations and HQ planning and processes related to the Global Refugee Forum have 
supported regionally to optimize realization of GCR and GRF objectives.
46,144.50
Partnerships with UN, NGOs, MDBs, IFIs and other development partners and other actors are 
strengthened and support to country operations is provided
46,144.50
Regional engagement of partners in a whole-of-society approach supports GRF development 
and pledge implementation.
46,144.50
Communications and media are effectively used at country and at RB levels to build awareness of, 
empathy with, and concrete support for refugees and other forcibly displaced persons in the East, 
Horn and Great Lakes Region in line with the bureau’s strategic objectives and to position UNHCR as 
an effective, impactful organization (EA20: External)
46,144.50
Media engagement and external communications are strengthened through support to country 
operations, effective coordination with HQs and the three regional bureaus in Africa, and 
production of quality external products reflecting prioritized advocacy aims and resource 
mobilization efforts.
46,144.50
Donors are engaged and kept informed of operational needs in the EHAGL region to mobilize 
resources and sustain support to refugees and persons of concern in the region (EA20: External)
92,289.00
Regional inputs to Global Appeals, Global Reports, Supplementary Appeals, Donor submissions, 
Concept Notes and Donor reports are delivered
46,144.50
Resource mobilization training and capacity development needs of the operations identified and 
addressed
46,144.50
Total
18,334,896.71















3.3 - Budget Overview - Staff OP
Current "Implementation - Draft" OP/OL budget as currently recorded in COMPASS



2022

OP
FTE
3. Operational Delivery
9,572,370.74
60.0
D1
290,290.32
1.0
G5
63,358.38

NA
68,667.53
1.0
NC
115,550.38
1.0
P4
873,053.76
4.0
P5
257,638.69
1.0
5. Administration & Finance
3,165,069.50
30.0
D1
580,580.64
2.0
D2
643,894.42
2.0
G2
102,214.35
7.0
G3
20,501.25
1.0
G4
25,547.73
1.0
G5
63,358.38
2.0
G6
301,163.28
8.0
P3
733,643.88
4.0
P4
436,526.88
2.0
P5
257,638.69
1.0
6. Supply
719,332.96
6.0
G4
25,547.73
1.0
G5
31,679.19
1.0
G6
37,645.41
1.0
P3
366,821.94
2.0
P5
257,638.69
1.0
7. Human Resources
691,184.34
6.0
G5
31,679.19
1.0
G6
75,290.82
2.0
P2
143,164.67
1.0
P3
183,410.97
1.0
P5
257,638.69
1.0
8. Information Technology
470,999.01
4.0
G5
31,679.19
1.0
G6
37,645.41
1.0
P3
183,410.97
1.0
P4
218,263.44
1.0
Total
16,274,896.71
115.0










Section 4: Monitoring and Evaluation

Monitoring and Evaluation content under review."""
])
```
